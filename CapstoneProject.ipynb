```python
import os
os.environ["TF_FORCE_GPU_ALLOW_GROWTH"] = "true"
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'

import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import collections
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import pickle
import seaborn as sns
import kagglehub

# Setup GPU
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Fungsi untuk memuat dataset dengan augmentasi yang lebih optimal
def load_dataset(base_path, image_size=(224, 224), batch_size=32, augment=False):
    def process_image(path, label):
        img = tf.io.read_file(path)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, [256, 256])
        img = tf.image.random_crop(img, size=[image_size[0], image_size[1], 3])
        img = tf.cast(img, tf.float32) / 255.0

        if augment:
            # Augmentasi yang lebih efisien
            img = tf.image.random_flip_left_right(img)
            if tf.random.uniform(()) > 0.5:
                img = tf.image.random_brightness(img, max_delta=0.1)
                img = tf.image.random_contrast(img, lower=0.9, upper=1.1)
            img = tf.clip_by_value(img, 0.0, 1.0)

        return img, label

    image_paths = []
    labels = []
    for root, dirs, files in os.walk(base_path):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(root, file))
                labels.append(os.path.basename(root))

    le = LabelEncoder()
    labels_encoded = le.fit_transform(labels)

    ds = tf.data.Dataset.from_tensor_slices((image_paths, labels_encoded))
    ds = ds.shuffle(len(image_paths)) if augment else ds
    ds = ds.map(process_image, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(batch_size)
    ds = ds.prefetch(tf.data.AUTOTUNE)

    return ds, le

# Arsitektur CNN yang lebih cepat dengan Depthwise Separable Convolution
def build_fast_cnn(input_shape=(224, 224, 3), num_classes=2):
    inputs = tf.keras.Input(shape=input_shape)

    # Stem layer yang efisien
    x = layers.Conv2D(32, (3, 3), strides=(2, 2), padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)

    # Blok konvolusi terpisah untuk efisiensi
    def depthwise_block(x, filters, strides=1):
        # Depthwise convolution
        x = layers.DepthwiseConv2D((3, 3), strides=strides, padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)

        # Pointwise convolution
        x = layers.Conv2D(filters, (1, 1), padding='same')(x)
        x = layers.BatchNormalization()(x)
        return layers.Activation('relu')(x)

    # Blok utama
    x = depthwise_block(x, 64)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Dropout(0.3)(x)

    x = depthwise_block(x, 128, strides=2)
    x = layers.Dropout(0.3)(x)

    x = depthwise_block(x, 256, strides=2)
    x = layers.GlobalAveragePooling2D()(x)

    # Head network
    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    return tf.keras.Model(inputs, outputs)

# Fungsi untuk menghitung class weights
def calculate_class_weights(labels):
    class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
    return dict(enumerate(class_weights))

# Training loop yang dioptimasi
def train_model(model, train_ds, val_ds, epochs, class_weights):
    # Konversi class weights ke tensor
    class_weight_values = tf.constant(list(class_weights.values()), dtype=tf.float32)

    # Metrik
    train_loss_metric = tf.keras.metrics.Mean(name='train_loss')
    train_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy(name='train_acc')
    val_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy(name='val_acc')

    # Optimizer dengan learning rate schedule
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=0.001,
        decay_steps=1000,
        decay_rate=0.9)
    optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)

    # Loss function
    loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()

    history = {
        'loss': [],
        'accuracy': [],
        'val_loss': [],
        'val_accuracy': []
    }

    @tf.function
    def train_step(images, labels):
        with tf.GradientTape() as tape:
            predictions = model(images, training=True)
            loss = loss_fn(labels, predictions)
            weights = tf.gather(class_weight_values, labels)
            loss = tf.reduce_mean(loss * weights)
            loss += sum(model.losses)

        gradients = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))

        train_loss_metric.update_state(loss)
        train_acc_metric.update_state(labels, predictions)
        return loss

    for epoch in tqdm(range(epochs), desc="Training", unit="epoch"):
        train_loss_metric.reset_state()
        train_acc_metric.reset_state()
        val_acc_metric.reset_state()

        # Training
        for images, labels in tqdm(train_ds, desc=f"Epoch {epoch+1}/{epochs}", leave=False):
            train_step(images, labels)

        # Validation
        val_loss = []
        for val_images, val_labels in val_ds:
            val_preds = model(val_images, training=False)
            val_loss.append(loss_fn(val_labels, val_preds).numpy())
            val_acc_metric.update_state(val_labels, val_preds)

        # Record metrics
        history['loss'].append(train_loss_metric.result().numpy())
        history['accuracy'].append(train_acc_metric.result().numpy())
        history['val_loss'].append(np.mean(val_loss))
        history['val_accuracy'].append(val_acc_metric.result().numpy())

        print(f"\nEpoch {epoch+1}: "
              f"Train Loss: {history['loss'][-1]:.4f}, "
              f"Train Acc: {history['accuracy'][-1]:.4f}, "
              f"Val Loss: {history['val_loss'][-1]:.4f}, "
              f"Val Acc: {history['val_accuracy'][-1]:.4f}")

    return history

# Fungsi untuk plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Main execution
def main():
    # Download dataset
    print("Downloading dataset...")
    try:
        dataset_path = kagglehub.dataset_download("techsash/waste-classification-data")
        print(f"Dataset path: {dataset_path}")
    except Exception as e:
        print(f"Failed to download dataset: {e}")
        return

    # Define paths
    train_path = os.path.join(dataset_path, "DATASET/TRAIN")
    test_path = os.path.join(dataset_path, "DATASET/TEST")

    # Verify paths
    if not os.path.exists(train_path):
        print(f"Error: Train directory not found at {train_path}")
        return

    # Load data
    train_ds, le = load_dataset(train_path, augment=True, batch_size=64)
    test_ds, _ = load_dataset(test_path, batch_size=64)

    # Split into train and validation
    train_size = int(0.8 * len(train_ds))
    val_ds = train_ds.skip(train_size)
    train_ds = train_ds.take(train_size)

    # Calculate class weights
    print("Calculating class weights...")
    all_labels = np.concatenate([y for x, y in train_ds], axis=0)
    class_weights = calculate_class_weights(all_labels)
    print("Class weights:", class_weights)

    # Build model
    print("Building model...")
    model = build_fast_cnn(num_classes=len(le.classes_))
    model.summary()

    # Train model
    print("Starting training...")
    history = train_model(
        model=model,
        train_ds=train_ds,
        val_ds=val_ds,
        epochs=30,
        class_weights=class_weights
    )

    # Evaluate on test set
    print("Evaluating on test set...")
    y_true = []
    y_pred = []
    for images, labels in tqdm(test_ds, desc="Evaluating"):
        predictions = model.predict(images, verbose=0)
        y_true.extend(labels.numpy())
        y_pred.extend(np.argmax(predictions, axis=1))

    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=le.classes_))

    # Plot confusion matrix
    plot_confusion_matrix(y_true, y_pred, le.classes_)

    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Train Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Save model
    model.save('/content/drive/MyDrive/CP/waste_classifier_fast_cnn_final.h5')
    print("Model saved to Google Drive as 'waste_classifier_fast_cnn.h5'")

    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    tflite_model = converter.convert()
    with open('/content/drive/MyDrive/CP/waste_classifier_fast_cnn_final.tflite', 'wb') as f:
        f.write(tflite_model)

    # Save label encoder
    with open('/content/drive/MyDrive/CP/label_encoder.pkl', 'wb') as f:
        pickle.dump(le, f)
    print("Label encoder saved to Google Drive as 'label_encoder.pkl'")

if __name__ == "__main__":
    main()
```

# CNN DENGAN DEPTHWISE SEPARABLE CONVOLUTION

**Input:** `224x224x3`

---

**Conv2D** `3x3`, filters=32, stride=2, padding='same'  
→ BatchNorm → ReLU  
**Output:** `112x112x32`

---

### Depthwise Block 1
- DepthwiseConv2D `3x3` → BatchNorm → ReLU  
- PointwiseConv2D `1x1`, filters=64 → BatchNorm → ReLU  
**Output:** `112x112x64`

---

**MaxPooling2D** `2x2` → Dropout(0.3)  
**Output:** `56x56x64`

---

### Depthwise Block 2 (stride=2)
- DepthwiseConv2D `3x3`, stride=2 → BatchNorm → ReLU  
- PointwiseConv2D `1x1`, filters=128 → BatchNorm → ReLU  
→ Dropout(0.3)  
**Output:** `28x28x128`

---

### Depthwise Block 3 (stride=2)
- DepthwiseConv2D `3x3`, stride=2 → BatchNorm → ReLU  
- PointwiseConv2D `1x1`, filters=256 → BatchNorm → ReLU  
**Output:** `14x14x256`

---

**GlobalAveragePooling2D()**  
**Output:** `256`

---

**Dense** `128`, activation=ReLU, `kernel_regularizer=l2(0.01)`  
→ Dropout(0.5)  
**Output:** `128`

---

**Dense** `num_classes`, activation=softmax  
**Output:** `num_classes`

```python
import os
os.environ["TF_FORCE_GPU_ALLOW_GROWTH"] = "true"
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'

import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import collections
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import pickle
import seaborn as sns
import kagglehub

# Setup GPU
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Fungsi untuk memuat dataset dengan augmentasi yang lebih optimal
def load_dataset(base_path, image_size=(224, 224), batch_size=32, augment=False):
    def process_image(path, label):
        img = tf.io.read_file(path)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, [256, 256])
        img = tf.image.random_crop(img, size=[image_size[0], image_size[1], 3])
        img = tf.cast(img, tf.float32) / 255.0

        if augment:
            img = tf.image.random_flip_left_right(img)
            if tf.random.uniform(()) > 0.5:
                img = tf.image.random_brightness(img, max_delta=0.1)
                img = tf.image.random_contrast(img, lower=0.9, upper=1.1)
            img = tf.clip_by_value(img, 0.0, 1.0)

        return img, label

    image_paths = []
    labels = []
    for root, dirs, files in os.walk(base_path):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(root, file))
                labels.append(os.path.basename(root))

    le = LabelEncoder()
    labels_encoded = le.fit_transform(labels)

    ds = tf.data.Dataset.from_tensor_slices((image_paths, labels_encoded))
    ds = ds.shuffle(len(image_paths)) if augment else ds
    ds = ds.map(process_image, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(batch_size)
    ds = ds.prefetch(tf.data.AUTOTUNE)

    return ds, le

# Arsitektur CNN yang lebih cepat dengan Depthwise Separable Convolution
def build_fast_cnn(input_shape=(224, 224, 3), num_classes=2):
    inputs = tf.keras.Input(shape=input_shape)

    x = layers.Conv2D(32, (3, 3), strides=(2, 2), padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)

    def depthwise_block(x, filters, strides=1):
        x = layers.DepthwiseConv2D((3, 3), strides=strides, padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        x = layers.Conv2D(filters, (1, 1), padding='same')(x)
        x = layers.BatchNormalization()(x)
        return layers.Activation('relu')(x)

    x = depthwise_block(x, 64)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Dropout(0.3)(x)

    x = depthwise_block(x, 128, strides=2)
    x = layers.Dropout(0.3)(x)

    x = depthwise_block(x, 256, strides=2)
    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    return tf.keras.Model(inputs, outputs)

def calculate_class_weights(labels):
    class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
    return dict(enumerate(class_weights))

def get_last_checkpoint_epoch(checkpoint_dir):
    if not os.path.exists(checkpoint_dir):
        return 0
    checkpoints = [f for f in os.listdir(checkpoint_dir) if f.startswith('cp-') and f.endswith('.weights.h5')]
    if not checkpoints:
        return 0
    epochs = []
    for f in checkpoints:
        try:
            epoch_num = int(f.split('-')[1].split('.')[0])
            epochs.append(epoch_num)
        except (IndexError, ValueError):
            continue
    return max(epochs) if epochs else 0

class SaveBestAccuracy(tf.keras.callbacks.Callback):
    def __init__(self, filepath, monitor='val_accuracy', mode='max'):
        super().__init__()
        self.filepath = filepath
        self.monitor = monitor
        self.mode = mode
        self.best = -np.inf if mode == 'max' else np.inf

        # Load previous best if exists
        if os.path.exists(filepath):
            with open(filepath, 'r') as f:
                self.best = float(f.read().strip())

    def on_epoch_end(self, epoch, logs=None):
        current = logs.get(self.monitor)
        if current is None:
            return
        if (self.mode == 'max' and current > self.best) or (self.mode == 'min' and current < self.best):
            self.best = current
            with open(self.filepath, 'w') as f:
                f.write(str(current))

class CustomModelCheckpoint(tf.keras.callbacks.Callback):
    def __init__(self, filepath, monitor='val_accuracy', mode='max'):
        super().__init__()
        self.filepath = filepath
        self.monitor = monitor
        self.mode = mode
        self.best = -np.inf if mode == 'max' else np.inf
        self.best_file = filepath.replace('.keras', '.txt')

        if os.path.exists(self.best_file):
            try:
                with open(self.best_file, 'r') as f:
                    self.best = float(f.read().strip())
            except:
                pass

    def on_epoch_end(self, epoch, logs=None):
        current = logs.get(self.monitor)
        if current is None:
            return

        improved = (self.mode == 'max' and current > self.best) or (self.mode == 'min' and current < self.best)
        if improved:
            self.best = current
            print(f"\nNew best {self.monitor}: {current:.4f}, saving model to {self.filepath}")
            self.model.save(self.filepath)
            with open(self.best_file, 'w') as f:
                f.write(str(current))
        else:
            print(f"\n{self.monitor} did not improve (current: {current:.4f}, best: {self.best:.4f})")

def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

def main():
    print("Downloading dataset...")
    try:
        dataset_path = kagglehub.dataset_download("techsash/waste-classification-data")
        print(f"Dataset path: {dataset_path}")
    except Exception as e:
        print(f"Failed to download dataset: {e}")
        return

    train_path = os.path.join(dataset_path, "DATASET/TRAIN")
    test_path = os.path.join(dataset_path, "DATASET/TEST")

    if not os.path.exists(train_path):
        print(f"Error: Train directory not found at {train_path}")
        return

    train_ds, le = load_dataset(train_path, augment=True, batch_size=64)
    test_ds, _ = load_dataset(test_path, batch_size=64)

    train_size = int(0.8 * len(train_ds))
    val_ds = train_ds.skip(train_size)
    train_ds = train_ds.take(train_size)

    print("Calculating class weights...")
    all_labels = np.concatenate([y.numpy() for _, y in train_ds], axis=0)
    class_weights = calculate_class_weights(all_labels)
    print("Class weights:", class_weights)

    print("Building model...")
    model = build_fast_cnn(num_classes=len(le.classes_))
    model.summary()

    checkpoint_dir = '/content/drive/MyDrive/CP/training_checkpoints'
    initial_epoch = get_last_checkpoint_epoch(checkpoint_dir)

    if initial_epoch > 0:
        print(f"Resuming training from epoch {initial_epoch}")
        best_model_path = os.path.join(checkpoint_dir, "best_model.keras")
        if os.path.exists(best_model_path):
            print(f"Loading best model from {best_model_path}")
            model.load_weights(best_model_path)
        else:
            print("Best model not found. Loading latest checkpoint.")
            latest_checkpoint = os.path.join(checkpoint_dir, f"cp-{initial_epoch:04d}.weights.h5")
            model.load_weights(latest_checkpoint)
    else:
        print("Starting new training")

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        metrics=['accuracy']
    )

    callbacks = [
        tf.keras.callbacks.ModelCheckpoint(
            filepath=os.path.join(checkpoint_dir, "cp-{epoch:04d}.weights.h5"),
            save_weights_only=True,
            verbose=1
        ),
        CustomModelCheckpoint(
            filepath=os.path.join(checkpoint_dir, "best_model.keras"),
            monitor='val_accuracy',
            mode='max'
        ),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=15,
            mode='max',
            restore_best_weights=True
        )
    ]

    best_acc_path = os.path.join(checkpoint_dir, "best_accuracy.txt")
    if os.path.exists(best_acc_path):
        with open(best_acc_path, 'r') as f:
            try:
                prev_best_acc = float(f.read().strip())
                print(f"Previous best val_accuracy: {prev_best_acc:.4f}")
            except:
                pass

    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=100,
        initial_epoch=initial_epoch,
        class_weight=class_weights,
        callbacks=callbacks
    )

    best_model_path = os.path.join(checkpoint_dir, "best_model.keras")
    if os.path.exists(best_model_path):
        print(f"Loading best model from {best_model_path} for evaluation...")
        model = tf.keras.models.load_model(best_model_path)

    print("Evaluating on test set...")
    y_true = []
    y_pred = []
    for images, labels in tqdm(test_ds, desc="Evaluating"):
        predictions = model.predict(images, verbose=0)
        y_true.extend(labels.numpy())
        y_pred.extend(np.argmax(predictions, axis=1))

    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=le.classes_))
    plot_confusion_matrix(y_true, y_pred, le.classes_)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

    model.save('/content/drive/MyDrive/CP/waste_classifier_fast_cnn_final.h5')
    print("Final model saved as 'waste_classifier_fast_cnn_final.h5'")

    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    tflite_model = converter.convert()
    with open('/content/drive/MyDrive/CP/waste_classifier_fast_cnn_final.tflite', 'wb') as f:
        f.write(tflite_model)

    with open('/content/drive/MyDrive/CP/label_encoder.pkl', 'wb') as f:
        pickle.dump(le, f)
    print("Label encoder saved as 'label_encoder.pkl'")

if __name__ == "__main__":
    main()
```
